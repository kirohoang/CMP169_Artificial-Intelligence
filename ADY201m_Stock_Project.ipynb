```python
# Lấy dữ liệu cổ phiếu TSLA từ 12/09/2024 đến 12/09/2024
import requests
import pandas as pd

# URL của API
url = 'https://api.nasdaq.com/api/quote/TSLA/historical?assetclass=stocks&fromdate=2019-09-12&limit=9999&todate=2024-09-12'

# Gửi request tới API
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
data = response.json()

# Trích xuất dữ liệu từ JSON
rows = data['data']['tradesTable']['rows']

# Chuyển dữ liệu thành DataFrame
df_tsla = pd.DataFrame(rows)
```

## 3. Data preprocessing - 1 point
- Understand the data: what does the data describe, what do the data columns describe?
- What data has been preprocessed and how?

#### Xử lý dữ liệu TSLA

```python
import pandas as pd

# Viết hoa chữ cái đầu tên cột
df_tsla.columns = df_tsla.columns.str.capitalize()

# Chuyển cột date sang datetime
df_tsla['Date'] = pd.to_datetime(df_tsla['Date']) 

# Sắp xếp theo giời gian
df_tsla = df_tsla.sort_values(by='Date').reset_index(drop=True) 

# Xóa $ của các cột giá và chuyển sang float
cols = ['Close', 'Open', 'High', 'Low']
for col in cols:
    df_tsla[col] = df_tsla[col].str.replace('$', '', regex=False).astype(float)

# Xóa dấu phẩy cột Volume
df_tsla['Volume'] = df_tsla['Volume'].str.replace(",", "") 

# Chuyển cột Volume sang numberic
df_tsla['Volume'] = pd.to_numeric(df_tsla['Volume']) 

# Đưa cột Volume về cuối 
df_tsla['Volume'] = df_tsla.pop('Volume')

# Tạo cột % thay đổi của giá close
df_tsla['% Change'] = (df_tsla['Close'].pct_change() * 100).round(2) 

# Hàng đầu tiên giá trị trống nên tự tính toán (-0,5%)
df_tsla = df_tsla.fillna(-0.50) 

df_tsla
```

- Dữ liệu ghi lại biến động giá và khối lượng giao dịch hàng ngày của cổ phiếu Tesla (TSLA). Mỗi dòng đại diện cho một ngày giao dịch, với các mức giá phản ánh hoạt động cổ phiếu Tesla trong ngày, và khối lượng giao dịch thể hiện tổng số lượng cổ phiếu được giao dịch.

* **Close (Giá Đóng Cửa):**
Giá cổ phiếu vào cuối ngày giao dịch.
Mức giá được theo dõi nhiều nhất vì nó phản ánh giá cuối cùng được thỏa thuận trong các giao dịch của ngày đó.

* **Open (Giá Mở Cửa):**
Giá cổ phiếu vào lúc bắt đầu ngày giao dịch là thời điểm diễn ra giao dịch đầu tiên của ngày.

* **High (Giá Cao Nhất):**
Giá cao nhất mà cổ phiếu Tesla đạt được trong ngày giao dịch.

* **Low (Giá Thấp Nhất):**
Giá thấp nhất mà cổ phiếu Tesla đạt được trong ngày giao dịch.

* **Volume (Khối Lượng Giao Dịch):**
Tổng số cổ phiếu được giao dịch trong ngày đó.


#### Đọc dữ liệu lãi suất FED từ 12/09/2019 đến 12/09/2024

```python
df_fed = pd.read_csv('FED_DFF.csv')

# Chuyển Date sang datetime
df_fed['Date'] = pd.to_datetime(df_fed['Date'])

# Sắp xếp theo giời gian
df_fed = df_fed.sort_values(by='Date').reset_index(drop=True) 

# Chuyển cột lãi suất về float
df_fed['DFF'] = df_fed['DFF'].astype(float)

# Lọc dữ liệu FED để trùng ngày với dữ liệu TSLA 
df_fed_tsla = df_fed[df_fed['Date'].isin(df_tsla['Date'])].reset_index(drop=True)

df_fed_tsla
```

* DFF:
Lãi suất quỹ liên bang (Federal Funds Rate - DFF).
là lãi suất mà các ngân hàng thương mại tại Hoa Kỳ sử dụng để cho nhau vay các khoản tiền ngắn hạn, thường là qua đêm. là công cụ mà Fed sử dụng để kiểm soát mức cung tiền trong nền kinh tế và kiểm soát lạm phát. Bằng cách tăng hoặc giảm lãi suất này, Fed có thể ảnh hưởng đến hoạt động cho vay, tiêu dùng và đầu tư trong nền kinh tế. Lãi suất quỹ liên bang là một yếu tố quan trọng trong chính sách tiền tệ của Hoa Kỳ và có tác động sâu rộng đến nền kinh tế toàn cầu, bao gồm cả thị trường tài chính, lãi suất cho vay và mức độ đầu tư.

#### Đọc dữ liệu giá dầu Crude Oil WTI Futures từ 12/09/2019 đến 12/09/2024


```python
df_oil = pd.read_csv("Crude Oil WTI Futures.csv")

# Chuyển Date sang datetime
df_oil['Date'] = pd.to_datetime(df_oil['Date'])

# Sắp xếp theo giời gian
df_oil = df_oil.sort_values(by='Date').reset_index(drop=True) 

# Chuyển cột giá dầu về float
df_oil['Price'] = df_oil['Price'].astype(float)

# Lọc dữ liệu giá dầu để trùng ngày với dữ liệu TSLA 
df_oil_tsla = df_oil[df_oil['Date'].isin(df_tsla['Date'])].reset_index(drop=True)

df_oil_tsla
```

- Dữ liệu ghi lại biến động giá và khối lượng giao dịch hàng ngày của giá dầu thô WTI.  Khi giá dầu thô tăng, chi phí xăng dầu cũng tăng theo, điều này có thể khiến người tiêu dùng quan tâm hơn đến các phương tiện sử dụng năng lượng sạch và tiết kiệm như xe điện. #### Đọc dữ liệu lợi nhuận của TESLA từ Quý 3 2019 đến Quý 2 2024


```python
# df_revenue = pd.read_excel('tsla_revenue.xlsx', header=None)
df_revenue = pd.read_csv('tsla_revenue.csv', header=None)

# Đặt tên cột cho df_revenue
df_revenue.columns = ['Date', 'Revenue']

# Chuyển Date sang datetime
df_revenue['Date'] = pd.to_datetime(df_revenue['Date'])

# Sắp xếp theo giời gian
df_revenue = df_revenue.sort_values(by='Date').reset_index(drop=True) 

# Thêm cột thời gian theo Quý
df_revenue['Quarter'] = df_revenue['Date'].dt.to_period('Q')

# Xóa '$' và 'B' của các cột Revenue và chuyển sang float
df_revenue['Revenue'] = df_revenue['Revenue'].str.replace('$', '', regex=False).str.replace('B', '', regex=False).astype(float)

df_revenue
```

Revenue: Dữ liệu doanh thu của Tesla tính theo từng quý. 
- Dữ liệu cho thấy doanh thu hàng quý của một công ty từ quý 3 năm 2019 đến quý 2 năm 2024. Doanh thu tăng đều đặn từ năm 2019 đến năm 2022, đạt đỉnh điểm vào quý 4 năm 2022. Có một sự sụt giảm nhẹ vào quý 1 năm 2023, nhưng doanh thu đã phục hồi và duy trì ở mức cao trong phần còn lại của năm. Quý 1 năm 2024 chứng kiến doanh thu thấp hơn so với quý 4 năm 2023, nhưng quý 2 năm 2024 dự kiến sẽ phục hồi.

```python
import seaborn as sns

# Node lấy dữ liệu của cổ phiếu Tesla, lãi suất DFF và giá dầu đến ngày 01/07/2022
df_2022 = df_combined[['Close', 'DFF', 'Oil Price']].loc[:'2022-07-01']
# Node lọc dữ liệu doanh thu đến ngày 01/07/2022
df_revenue_2022 = df_revenue.loc[(df_revenue['Date'] <= '2022-07-01')]
# Node khởi tạo dataframe cho heatmap với các cột 'Date' và 'Revenue'
df_heatmap = df_revenue_2022[['Date', 'Revenue']]

# Thêm các cột 'Tesla', 'DFF' và 'Oil' với giá trị khởi tạo là 0.0
df_heatmap['Tesla'] = 0.0
df_heatmap['DFF'] = 0.0
df_heatmap['Oil'] = 0.0

# Node tính giá trị trung bình động cho từng ngày và gán vào các cột 'Tesla', 'DFF' và 'Oil'
for i in range(len(df_revenue_2022)):
    # Lấy ngày tại dòng thứ i trong df_revenue_2022
    date = df_revenue_2022['Date'].iloc[i]
    # Tính trung bình của cột 'Close' đến ngày đó và gán vào cột 'Tesla'
    df_heatmap.loc[i, 'Tesla'] = df_2022['Close'].loc[:date].mean()
    # Tính trung bình của cột 'DFF' đến ngày đó và gán vào cột 'DFF'
    df_heatmap.loc[i, 'DFF'] = df_2022['DFF'].loc[:date].mean()
    # Tính trung bình của cột 'Oil Price' đến ngày đó và gán vào cột 'Oil'
    df_heatmap.loc[i, 'Oil'] = df_2022['Oil Price'].loc[:date].mean()

# Node vẽ biểu đồ heatmap để hiển thị mối quan hệ tương quan giữa các cột
sns.heatmap(df_heatmap[df_heatmap.columns[1:]].corr(), annot=True, cmap='coolwarm')
```

Phân tích Heatmap về độ tương quan giữa các yếu tố ảnh hưởng đến giá cổ phiếu Tesla giai đoạn tháng 9/2019 đến tháng 7/2022:  
- Ta thấy lợi nhuận và giá cổ phiếu có độ tương quan thuận rất cao (khoảng 96%), còn lãi suất liên bang và giá cổ phiếu có độ tương quan nghịch cũng rất cao (-88%).
